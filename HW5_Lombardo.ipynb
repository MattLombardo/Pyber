```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)
```

```python
#city_data.head()
```

```python
#ride_data.head()
```

```python
combined_data = pd.merge(city_data, ride_data, how="outer", on="city")

combined_data.head()
```

```python
city_df = pd.DataFrame({"Avg Fare" : combined_data.groupby("city")["fare"].mean(),
                        "Total Rides" : combined_data["city"].value_counts(),
                        "Total Drivers" : combined_data.groupby("city")["driver_count"].mean(),
                        "Type" : city_data.set_index("city")["type"]})
city_df.head()
```

```python
urban_df = city_df.loc[city_df["Type"] == "Urban"]
rural_df = city_df.loc[city_df["Type"] == "Rural"]
suburban_df = city_df.loc[city_df["Type"] == "Suburban"]

#urban_df
#rural_df
#suburban_df
```

# Bubble Plot

```python
plt.figure(figsize=(10,7))
plt.grid(True)

plt.scatter(urban_df["Total Rides"], urban_df["Avg Fare"], linewidth=1.3,
            marker="o", facecolors="lightcoral", edgecolors="black", 
            s=urban_df["Total Drivers"]*20, alpha=0.6, label="Urban")
plt.scatter(rural_df["Total Rides"], rural_df["Avg Fare"], linewidth=1.3, 
            marker="o", facecolors="gold", edgecolors="black", 
            s=rural_df["Total Drivers"]*20, alpha=0.6, label="Rural")
plt.scatter(suburban_df["Total Rides"], suburban_df["Avg Fare"], linewidth=1.3,
            marker="o", facecolors="lightskyblue", edgecolors="black", 
            s=suburban_df["Total Drivers"]*20, alpha=0.6, label="Suburban")

plt.scatter([],[], s=0, label=" ")
plt.scatter([],[], s=0, label="Note: Circle size correlates to")
plt.scatter([],[], s=0, label="          total drivers (per city)")

plt.title("Pyber Ride Sharing Data")
plt.xlabel("Total Rides (per city)")
plt.ylabel("Average Fare (in dollars)")
plot_legend = plt.legend(title="City Type", loc=1)
for i in range(0,3):
    plot_legend.legendHandles[i]._sizes = [100]

plt.show()
```

# Pie Charts

### 1) Total Fares by City Type

```python
labels1 = ["Urban", "Suburban", "Rural"]
counts1 = [combined_data.loc[combined_data["type"] == "Urban"]["fare"].sum(),
          combined_data.loc[combined_data["type"] == "Suburban"]["fare"].sum(),
          combined_data.loc[combined_data["type"] == "Rural"]["fare"].sum()]
colors1 = ["lightcoral", "lightskyblue", "gold"]
explode1 = (0.05, 0, 0)

plt.figure(figsize=(7,7))
plt.pie(counts1, explode=explode1, labels=labels1, colors=colors1, 
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.axis("equal")
plt.title("% of Total Fares by City Type")

plt.show()
```

### 2) Total Rides by City Type

```python
labels2 = ["Urban", "Suburban", "Rural"]
counts2 = [urban_df["Total Rides"].sum(),
           suburban_df["Total Rides"].sum(),
           rural_df["Total Rides"].sum()]
colors2 = ["lightcoral", "lightskyblue", "gold"]
explode2 = (0.05, 0, 0)

plt.figure(figsize=(7,7))
plt.pie(counts2, explode=explode2, labels=labels2, colors=colors2, 
        autopct="%1.1f%%", shadow=True, startangle=70)
plt.axis("equal")
plt.title("% of Total Rides by City Type")

plt.show()
```

### 3) Total Drivers by City Type

```python
labels3 = ["Urban", "Suburban", "Rural"]
counts3 = [urban_df["Total Drivers"].sum(),
           suburban_df["Total Drivers"].sum(),
           rural_df["Total Drivers"].sum()]
colors3 = ["lightcoral", "lightskyblue", "gold"]
explode3 = (0.07, 0, 0)

plt.figure(figsize=(7,7))
plt.pie(counts3, explode=explode3, labels=labels3, colors=colors3, 
        autopct="%1.1f%%", shadow=True, startangle=180)
plt.axis("equal")
plt.title("% of Total Drivers by City Type")

plt.show()
```

## Three observable trends based on the data:
* When looking at the bubble plot, the data follows a fairly clear trend that the rural city type has high average fares and low total rides/ drivers, the suburban city type has medium average fares and medium total rides/ drivers, and the urban city type has low average fares and high total rides/ drivers. This trend logically makes sense since an urban city type has a very high amount of people living there and thus has a high number of drivers and a high number of rides, however since everything is so close together the average fare is low because it does not take long to get to your destination. Conversely, a rural city type has a low amount of people living there compared to the urban city type and thus has a low number of drivers and low number of rides by comparison. However, since the rural city type is less geographically dense and everything is much more spread out, the average fare is high because it takes longer to drive to your destination as compared to the urban city type. 

* When looking at the pie charts some interesting subtle trends can be observed. While the urban city type dominates the percent of total drivers at 80.9%, its percent of total rides drops off by a considerable amount at 68.4%. With such a large percent of the total drivers you would imagine that its percent of total rides would be equally large, however it is over 10% lower. Since the percent of total rides is significantly lower than the percent of total drivers, you can assume that there is a saturation of drivers in the urban city type and because of that they are struggling to all get rides/ work. With an even lower percent of total fares at 62.7%, it appears that being a driver in the urban city type is tough because proportionally the fares are low and all of the competition leads to a drought of rides/ work.

* Connected with the observation above, the opposite trend can be seen with the suburban city type. Its percent of total drivers (16.5%) is considerably less than its percent of total rides (26.3%), signifying that there is probably a lack of drivers in the suburban city type as compared to the demand for rides. Because of this, drivers in the suburban city type have a plethora of ride opportunities and thus have plenty of work. With an even higher percent of total fares (30.5%), this signifies that being a driver in the suburban city type is much more lucrative because there are plenty of rides and the fares are proportionally good.
